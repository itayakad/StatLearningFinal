```python
import json
import requests
import pandas as pd
```

```python

def get_news(start,end,stocks = ['AAPL']):
    stock_str = "%2C".join(stocks)
    url = f"https://data.alpaca.markets/v1beta1/news?start={start}&end={end}&sort=asc&symbols={stock_str}&limit=50&include_content=false"
    print(url)
    headers = {"accept": "application/json",
        "APCA-API-KEY-ID": "PKWM6ASMJDOXR7U7BFXV",
        "APCA-API-SECRET-KEY": "GeOgITwNNHx6IcaAeWE8VGYzmGjSvxwwvug50Ofm"
    }
    response = requests.get(url, headers=headers)
    return json.loads(response.text)
```

```python
import pandas as pd
from datetime import datetime, timedelta

start_date = datetime.strptime("2025-01-01", "%Y-%m-%d")
end_date = datetime.strptime("2025-05-01", "%Y-%m-%d")
output_file = "daily_news.csv"

# Keep track of whether we've written the header
header_written = False

current = start_date
while current <= end_date:
    next_day = current + timedelta(days=1)
    date_str = current.strftime("%Y-%m-%d")
    next_str = next_day.strftime("%Y-%m-%d")

    news = get_news(date_str, next_str, stocks=['AAPL', 'GOOGL', 'NVDIA'])

    if news['news']:
        df = pd.DataFrame(news['news'])
        df["scrape_date"] = date_str  # add scrape date for tracking

        df.to_csv(output_file, mode='a', index=False, header=not header_written)
        header_written = True
        print(f"Wrote news for {date_str}")

    current = next_day
```
